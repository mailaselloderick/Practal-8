```python
import pandas as pd
```

```python
df_ramen=pd.read_csv('ramen-ratings.csv')
df_ramen.head()
```

```python
#1. Group the data and get the average of the groups to produce a DataFrame that looks like this:
grouped_cols=df_ramen.groupby(['Country','Brand','Style'])
average_col=grouped_cols[['Stars']].mean()
average_col.head()
```

```python
#2. Group the data in the same way, but this time get both the average and the count.
grouped_cols=df_ramen.groupby(['Country','Brand','Style'],observed=True)
aggregate_cols=grouped_cols[['Stars']].agg(['mean','count'])
aggregate_cols.head()
```

```python
#3.Create a pivot table that shows the average stars by ramen style for each country.
df_ramen.pivot_table(index='Country',columns='Style',values='Stars',aggfunc='mean').head()
```

```python
#4. Calculate the total number of stars for each country, and then 
# rank the countries in descending order by total stars. 
total_stars = df_ramen.groupby('Country')['Stars'].sum().reset_index()
total_stars_sorted = total_stars.sort_values(by='Stars', ascending=False)
total_stars_sorted.head()
```

```python
#5. Bin the stars data into 11 bins. Label the bins from 0.0 to 5.0 in increments of .5, and
# then display the count of the stars in each bin
# Cut the stars into bins
df_ramen['Stars_bin'] = pd.cut(df_ramen['Stars'], 
    bins=[0.0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,5.0,5.5], 
    labels=['0.0','0.5','1.0','1.5','2.0','2.5','3.0','3.5','4.0','4.5','5.0'])

# Display count per bin
df_ramen['Stars_bin'].value_counts().sort_index()

```

```python
#6. Bin the stars data into five quantiles. Give each bin a descriptive label, and 
# then display the count of the stars in each bin.  

df_ramen['Stars_quantile'] = pd.qcut(df_ramen['Stars'], q=5, 
            labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
df_ramen['Stars_quantile'].value_counts().sort_index()
```
